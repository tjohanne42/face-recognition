# class FaceAnalyse

```python
import os

class FaceAnalyse(object):
    """
    params :
        img_path -> Format accepted : path to image | cv2 image
        df_path ->  (OPTIONAL) Path to folder containing image database for face recognition
                    If valid_path : return in addition a pandas.DataFrame
        draw ->     If True :   draw rectangle on all faces, circle on eyes, line on mouths
                                Return in addition a cv2 image
    return : 

    """

    def __init__(self, img_path, db_path=False, draw=True):

        self.img = self._load_img(img_path)


        """ load retina faces"""
        self.faces = RetinaFace.detect_faces(self.img)
        tmp = []
        for key in self.faces.keys():
            tmp.append(self.faces[key])
        self.faces = tmp
        """ self.retina[x]["score", "facial_area", "landmarks", "left_eye", "nose", "mouth_right", "mouth_left"] """


        """ face recognition """
        if db_path and type(db_path) == str:
            for i in range(len(self.faces)):
                facial_area = self.faces[i]["facial_area"]
                width = facial_area[3] - facial_area[1]
                height = facial_area[2] - facial_area[0]
                percent = 5
                percent_threshold = 100
                while percent <= percent_threshold:
                    crop_img = self.img[facial_area[1]-int(width*percent/100):facial_area[3]+int(width*percent/100),
                                   facial_area[0]-int(height*percent/100):facial_area[2]+int(height*percent/100)]
                    try:
                        df = DeepFace.find(img_path=crop_img, db_path=db_path, detector_backend="retinaface")
                        break
                    except:
                        percent += 5
                self.faces[i]["recognition"] = df
        print(self.faces)

    def _load_img(self, img_path):
        if type(img_path) == str:
            """ load img with path to file """
            if not os.path.isfile(img_path):
                raise ValueError("img_path not valid : \"{}\"".format(img_path))
            img = cv2.imread(img_path)
            if img.size == 0:
                raise ValueError("img_path not valid : \"{}\"".format(img_path))
            return img

        elif type(img_path) == np.ndarray:
            """ img_path is cv2 image """
            if img.size == 0:
                raise ValueError("img_path cv2 image not valid")
            return img_path

        """ img_path type not valid """
        raise ValueError("img_path type not valid : \"{}\"\nmust be path to image (str) or cv2 image (numpy.ndarray)".format(type(img_path)))
```

```python
paths = [
#         "known/chris_hemsworth.jpg",
#         "known/jeremy_renner.jpg",
#         "unknown/test_chris.jpg",
#         "unknown/test_chris_2.jpg",
#         "unknown/test1.jpg",
        "unknown/test2.jpg",
#         "unknown/test3.jpg",
#         "unknown/test4.jpg",
]

for path in paths:
    img = cv2.imread(path)
    drawn_img = img.copy()
    obj = RetinaFace.detect_faces(img)
    for key in obj.keys():
        identity = obj[key]

        facial_area = identity["facial_area"]
        right_eye = identity["landmarks"]["right_eye"]
        left_eye = identity["landmarks"]["left_eye"]
        nose = identity["landmarks"]["nose"]
        mouth_right = identity["landmarks"]["mouth_right"]
        mouth_left = identity["landmarks"]["mouth_left"]
        
        width = facial_area[3] - facial_area[1]
        height = facial_area[2] - facial_area[0]
        percent = 5
        percent_threshold = 100
        while percent <= percent_threshold:
            crop_img = img[facial_area[1]-int(width*percent/100):facial_area[3]+int(width*percent/100),
                           facial_area[0]-int(height*percent/100):facial_area[2]+int(height*percent/100)]
            try:
                demography = DeepFace.analyze(crop_img, detector_backend="retinaface")
                df = DeepFace.find(img_path = crop_img, db_path = "known", detector_backend="retinaface")
                break
            except:
                percent += 5
        if percent > percent_threshold:
            crop_img = img[facial_area[1]:facial_area[3], facial_area[0]:facial_area[2]]
            plt.imshow(crop_img[:, :, :: -1])
            plt.show()
            print("couldn't find any statistics on this one")
        else:
            display(df)
            plt.imshow(crop_img[:, :, :: -1])
            plt.show()
            print("emotion", demography["dominant_emotion"])
            print("age", demography["age"])
            print("gender", demography["gender"])
            print("race", demography["dominant_race"])
        
        thickness = 2
        
        cv2.rectangle(drawn_img, (facial_area[2], facial_area[3]), (facial_area[0], facial_area[1]), (0, 255, 255), thickness)
        cv2.circle(drawn_img, (int(right_eye[0]), int(right_eye[1])), radius=thickness, color=(0, 0, 255), thickness=-1)
        cv2.circle(drawn_img, (int(left_eye[0]), int(left_eye[1])), radius=thickness, color=(0, 0, 255), thickness=-1)
        cv2.circle(drawn_img, (int(nose[0]), int(nose[1])), radius=thickness, color=(0, 0, 255), thickness=-1)
        cv2.line(drawn_img, (int(mouth_right[0]), int(mouth_right[1])), (int(mouth_left[0]), int(mouth_left[1])), (0, 0, 255), thickness=thickness)
        

    plt.imshow(drawn_img[:, :, :: -1])
    plt.show()
```
